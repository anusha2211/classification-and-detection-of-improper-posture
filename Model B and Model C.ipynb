```python
import numpy as np
import pandas as pd
```

## Preprocessing

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
readings = df.iloc[:,0:18].values
readings = pd.DataFrame(readings)
readings.head()
```

### Normalization

```python
#Normalizing sensor readings
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

n_readings=scaler.fit_transform(readings)
n_readings = pd.DataFrame(n_readings)
```

```python
n_readings.head()
```

```python
df.head()
```

```python
p_type = df['type']
activity = df['activity']
```

```python
n_readings['type'] = p_type
n_readings['activity'] = activity
```

```python
n_readings.head()
```

#### Converting 'type' to integral values

```python
data = n_readings
data['type'] = data['type'].replace(['h'], 1)
data['type'] = data['type'].replace(['uh'], 0)
data.head()
```

#### Converting 'activity' to integral values

```python
data['activity'] = data['activity'].replace(['sitting'], 0)
data['activity'] = data['activity'].replace(['standing'], 1)
data['activity'] = data['activity'].replace(['sleeping'], 2)
data.head()
```

```python
final_data = data.to_csv("final_data.csv", index=False)
```

#### Generating 'combined' column

```python
df2 = pd.read_csv('final_data.csv')
df2['combined'] = df2[str('type')].map(str) + df2[str('activity')].map(str)
```

```python
df2.head()
```

#### Encoding 'combined' column

```python
df2['combined'] = df2['combined'].replace(['10'], 0)
df2['combined'] = df2['combined'].replace(['11'], 1)
df2['combined'] = df2['combined'].replace(['12'], 2)
df2['combined'] = df2['combined'].replace(['00'], 3)
df2['combined'] = df2['combined'].replace(['01'], 4)
df2['combined'] = df2['combined'].replace(['02'], 5)

df2.head()
multi_class = df2.to_csv("multiclass_final.csv", index=False)
```

```python
multi_class = pd.read_csv("multiclass_final.csv")
```

```python
import seaborn as sns
sns.pairplot(multi_class, hue = 'combined')
```

## Model B (binary)

```python
df = pd.read_csv('multiclass_final.csv')
```

```python
df.head()
```

```python
df['combined'].value_counts()
```

```python
X = df.iloc[:,0:18].values
Y = df['type']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state = 4)
```

### SVM

```python
from sklearn import svm

svm_model = svm.SVC(kernel='rbf', gamma=1, C=10)
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)
y_pred 
```

```python
print("Traning Accuracy : ", svm_model.score(X_train, y_train))
print("Testing Accuracy : ", svm_model.score(X_test, y_test)) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import  confusion_matrix
print(confusion_matrix(y_test,y_pred))

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

FP = cm.sum(axis=0) - np.diag(cm) 
FN = cm.sum(axis=1) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)
FP = FP.astype(float)
FN = FN.astype(float)
TP = TP.astype(float)
TN = TN.astype(float)
# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
print("Sensitivity/ Recall: ", TPR)
# Specificity or true negative rate
TNR = TN/(TN+FP) 
print("Specificity: ", TNR)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(max_depth=3, criterion = 'gini', min_samples_leaf = 5, min_samples_split = 2)
dt_model.fit(X_train, y_train)
y_pred1 = dt_model.predict(X_test)
```

```python
print("Traning Accuracy : ", dt_model.score(X_train, y_train))
print("Testing Accuracy : ", dt_model.score(X_test, y_test)) 
```

```python
from sklearn.neighbors import KNeighborsClassifier    
knn_model= KNeighborsClassifier(n_neighbors=1, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
y_pred= knn_model.predict(X_test)
```

```python
print("Traning Accuracy : ", knn_model.score(X_train, y_train))
print("Testing Accuracy : ", knn_model.score(X_test, y_test)) 
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train,y_train)
y_pred2 = rf_model.predict(X_test)
```

```python
print("Traning Accuracy : ", rf_model.score(X_train, y_train))
print("Testing Accuracy : ", rf_model.score(X_test, y_test))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier    
knn_model= KNeighborsClassifier(n_neighbors=1, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
y_pred= knn_model.predict(X_test)
```
